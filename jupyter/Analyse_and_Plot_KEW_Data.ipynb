```python
import numpy as np

# Simulate the power measurement for Sept (no voltage was measured)
dataSept["V1[V]"] = 238.0
dataSept["V2[V]"] = 239.0
dataSept["V3[V]"] = 243.0
dataSept["P[W]"] = dataSept["V1[V]"]*(dataSept["A1[A]"] + dataSept["A2[A]"] + dataSept["A3[A]"])
dataSept["P1[W]"] = dataSept["V1[V]"]*dataSept["A1[A]"]
dataSept["P2[W]"] = dataSept["V2[V]"]*dataSept["A2[A]"]
dataSept["P3[W]"] = dataSept["V3[V]"]*dataSept["A3[A]"]

# Simulate 5 new ACs on (more conservative) September data
AC = 2200
dataSept["P[W]"] += 5*AC
dataSept["P1[W]"] += 1*AC
dataSept["P2[W]"] += 2*AC
dataSept["P3[W]"] += 2*AC
```

# Plot data

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-dark')


def make_subplots(data):

    plt.figure(figsize=(18, 16), dpi= 100)

    plt.subplot(4, 1, 1)
    plt.title("P1[W]")
    plt.plot(data["DateTime"], data["P1[W]"])
    plt.ylabel("Power in W")
    plt.grid(True)


    plt.subplot(4, 1, 2)
    plt.title("P2[W]")
    plt.plot(data["DateTime"], data["P2[W]"])
    plt.ylabel("Power in W")
    plt.grid(True)


    plt.subplot(4, 1, 3)
    plt.title("P3[W]")
    plt.plot(data["DateTime"], data["P3[W]"])
    plt.ylabel("Power in W")
    plt.grid(True)


    plt.subplot(4, 1, 4)
    plt.title("P[W]")
    plt.plot(data["DateTime"], data["P[W]"][:])
    plt.ylabel("Power in W")
    plt.grid(True)


    plt.show()
```

```python
make_subplots(dataSept)
```

```python
make_subplots(dataNov)
```

```python
def make_plots(data):
    plt.figure(figsize=(18, 16), dpi= 100)
    plt.plot(data["DateTime"], data["P1[W]"], label="P1[W]")
    plt.plot(data["DateTime"], np.ones(data["DateTime"].shape)*np.mean(data["P1[W]"]), label="avg P1[W]: {} kW".format(round(np.mean(data["P1[W]"])/1000, 2)), color="k")
    
    plt.plot(data["DateTime"], data["P2[W]"], label="P2[W]")
    plt.plot(data["DateTime"], np.ones(data["DateTime"].shape)*np.mean(data["P2[W]"]), label="avg P2[W]: {} kW".format(round(np.mean(data["P2[W]"])/1000, 2)), color="k")
    plt.plot(data["DateTime"], data["P3[W]"], label="P3[W]")
    plt.plot(data["DateTime"], np.ones(data["DateTime"].shape)*np.mean(data["P3[W]"]), label="avg P3[W]: {} kW".format(round(np.mean(data["P3[W]"])/1000, 2)), color="k")
    plt.plot(data["DateTime"], data["P[W]"], label="P[W]") 
    plt.plot(data["DateTime"], np.ones(data["DateTime"].shape)*np.mean(data["P[W]"]), label="avg P[W]: {} kW".format(round(np.mean(data["P[W]"])/1000, 2)), color="k")

    plt.legend()
    plt.grid(True)
    plt.ylabel("Power in W")
    plt.show()
```

```python
make_plots(dataSept)
```

```python
make_plots(dataNov)
```

# Do some analysis

```python
import numpy as np

def analysis(data):

    P_total = 60000.0
    mean_P_total_in_W = np.mean(data["P[W]"])
    print("Total\n")
    print("Average demand: {} kW +- {} kW, max average output {} kW    --->   {} % of avg capacity\n".format(round(np.mean(data["P[W]"])/1000, 2), round(np.std(data["P[W]"])/1000, 2), 0.7*P_total/1000, round(np.mean(data["P[W]"])/(0.7*P_total)*100, 2)))
    print("Peak demand: {} kW, max peak output {} kW          --->   {} % of peak capacity\n\n".format(round(np.max(data["P[W]"])/1000, 2), P_total/1000, round(max(data["P[W]"])/P_total*100, 2)))

    mean_P = []
    max_P = []
    std_P = []
    
    for i in range(3):
        mean_P.append(np.mean(data["P{}[W]".format(i+1)]))
        std_P.append(np.std(data["P{}[W]".format(i+1)]))
        max_P.append(np.max(data["P{}[W]".format(i+1)]))
        print("-------------------------------------\n")
        print("Phase {}\n".format(i+1))
        print("  Average demand: {} kW +- {} kW, max average output {} kW    --->   {} % of avg capacity\n".format(round(mean_P[i]/1000, 2), round(std_P[i]/1000, 2), 0.7*(P_total/3)/1000, round(mean_P[i]/(0.7*(P_total/3))*100, 2)))
        print("  Peak demand: {} kW, max peak output {} kW         --->   {} % of peak capacity".format(round(max_P[i]/1000, 2), (P_total/3)/1000, round(max_P[i]/(P_total/3)*100, 2)))
        print("-------------------------------------\n")
```

```python
analysis(dataSept)
```
