# Plot data

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-dark')



plt.figure(figsize=(18, 16), dpi= 100)

plt.subplot(4, 1, 1)
plt.title("P1[W]")
plt.plot(data["DateTime"], data["P1[W]"])
plt.ylabel("Power in W")
plt.grid(True)


plt.subplot(4, 1, 2)
plt.title("P2[W]")
plt.plot(data["DateTime"], data["P2[W]"])
plt.ylabel("Power in W")
plt.grid(True)


plt.subplot(4, 1, 3)
plt.title("P3[W]")
plt.plot(data["DateTime"], data["P3[W]"])
plt.ylabel("Power in W")
plt.grid(True)


plt.subplot(4, 1, 4)
plt.title("P[W]")
plt.plot(data["DateTime"], data["P[W]"][:])
plt.ylabel("Power in W")
plt.grid(True)


plt.show()
```

```python
plt.figure(figsize=(18, 16), dpi= 100)
plt.plot(data["DateTime"], data["P1[W]"], label="P1[W]")
plt.plot(data["DateTime"], data["P2[W]"], label="P2[W]")
plt.plot(data["DateTime"], data["P3[W]"], label="P3[W]")
plt.plot(data["DateTime"], data["P[W]"], label="P[W]") 
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(18, 16), dpi= 100)
plt.plot(data["DateTime"], data["P1[W]"], label="P1[W]")
plt.plot(data["DateTime"], data["P2[W]"], label="P2[W]")
plt.plot(data["DateTime"], data["P3[W]"], label="P3[W]")
plt.legend()
plt.grid(True)
plt.show()
```

# Do some analysis

```python
import numpy as np

P_total = 48000.0
mean_P_total_in_W = np.mean(data["P[W]"])
print("CAT DE65E0 maximum capacity:     {} kW or 60 kVA \n".format(P_total/1000))
print("Average total power: {} kW    --->   {} % of average total capacity\n".format(round(np.mean(data["P[W]"])/1000, 2), round(np.mean(data["P[W]"])/P_total*100, 2)))
print("Max total power: {} kW        --->   {} % of maximum total capacity\n\n".format(round(np.max(data["P[W]"])/1000, 2), round(max(data["P[W]"])/P_total*100, 2)))

mean_P = []
max_P = []

for i in range(3):
    mean_P.append(np.mean(data["P{}[W]".format(i+1)]))
    max_P.append(np.max(data["P{}[W]".format(i+1)]))
    print("-------------------------------------\n")
    print("Phase {} - maximum capacity:    {} kW or 20 kVA\n".format(i+1, P_total/(3 * 1000)))
    print("  Average total power: {} kW    --->   {} % of average total capacity\n".format(round(mean_P[i]/1000, 2), round(mean_P[i]/(P_total/3)*100, 2)))
    print("  Max total power: {} kW       --->   {} % of maximum total capacity".format(round(max_P[i]/1000, 2), round(max_P[i]/(P_total/3)*100, 2)))
    print("-------------------------------------\n")
```
